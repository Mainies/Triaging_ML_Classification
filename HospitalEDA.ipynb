As per https://www.health.qld.gov.au/clinical-practice/guidelines-procedures/service-delivery/cscf/hospitals/public a list of hospitals that are considered "Urban" or non-rural


```python
open('remote-area-map.png')
```

```python
urbanlist = ['Mater Hospital Brisbane Public', 'Logan Hospital', 'Redcliffe Hospital', 'Ipswich Hospital', 'Queen Elizabeth II Jubilee Hospital',  'The Prince Charles Hospital', 'Princess Alexandra Hospital', 'Royal Brisbane & Women\'s Hospital', 'Queensland Children\'s Hospital', 'Sunshine Coast University Hospital', 'Gold Coast University Hospital', 'Robina Hospital']
```

```python
df_hospital['category'] = df_hospital['Facility/HHS Desc'].apply(lambda row: "Urban" if row in urbanlist else "Rural")
```

```python
df_hospital = df_hospital[df_hospital['Facility/HHS Desc'] != "QUEENSLAND"]
print(df_hospital.shape)
```

```python
df_hospital.category.value_counts()
```

```python
print(df_hospital[df_hospital['category'] == 'Urban'].shape)
```

```python
rural_hosp = df_hospital[df_hospital['category'] == "Rural"]
urban_hosp = df_hospital[df_hospital['category'] == 'Urban']
```

```python
sns.pairplot(rural_hosp, hue = 'Triage Category')
```

```python
sns.pairplot(urban_hosp, hue = 'Triage Category')
```

```python
rural_hosp = rural_hosp[rural_hosp['Number of Attendances'] < 1000]
```

```python
figure = plt.figure(figsize=(10,8))
ax = plt.subplot()
plt.hist(urban_hosp['Patients Seen within clinically recommended times (%)'], alpha = 0.4, density=True, label = 'Urban')
plt.hist(rural_hosp['Patients Seen within clinically recommended times (%)'], alpha = 0.4, density= True, label = "Rural")
plt.title('Percentage of Patients Seen in a Clinically Timely Manner')
plt.xlabel("Percentage")
plt.ylabel("Density")
plt.legend()
plt.show()
plt.clf()
```

```python
plt.figure(figsize=(10,8))
plt.subplot(2, 1, 1)
sns.scatterplot(data = urban_hosp, y = 'Patients Seen within clinically recommended times (%)', x = 'Number of Attendances', hue = 'Triage Category')
plt.title("Urban Hopsitals")
plt.subplot(2, 1, 2)
sns.scatterplot(data = rural_hosp, y = 'Patients Seen within clinically recommended times (%)', x = 'Number of Attendances', hue = 'Triage Category')
plt.title("Rural Hospitals")
plt.subplots_adjust(top=1.1)
plt.show()
```
